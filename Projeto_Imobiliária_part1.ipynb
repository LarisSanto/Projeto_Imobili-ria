# **Remover registros inconsistentes.**

Alguns registros existentes na base de dados não fazem muito sentido, por exemplo:


*  Apartamentos que possuem valor de aluguel igual a 0;

Esses registros são inconsistentes, por isso será  removido da base de dados.

```python
df.query("Valor == 0 | Condominio == 0")
```

```python
registros_a_remover = df.query("Valor == 0 | Condominio == 0").index
```

# **Aplicar filtros**

Nesse momento, o time de ML deseja avaliar apenas imóveis em cenários específicos, como:

*  Apartamentos que possuem 1 quarto e aluguel menor que R$ 1200;

*  Apartamentos que possuem pelo menos 2 quartos, aluguel menor que R$ 3000 e área maior que 70 m².

Com base nisso, vou aplicar filtros para obter essas informações específicas.

```python
# 1 - Apartamentos que possuem 1 quarto e aluguel menor que R$ 1200.
```

```python
df["Quartos"] == 1
```

```python
selecao1 = df["Quartos"] == 1
df[selecao1] # Somente valores que possui 1 quarto.
```

```python
selecao2 = df["Valor"] < 1200
df[selecao2]
```

```python
selecao_final = selecao1 & selecao2
df[selecao_final]
```

```python
df_1 = df[selecao_final]
```

```python
# 2 - Apartamentos que possuem pelo menos 2 quartos, aluguel menor que R$ 3000 e área maior que 70 m².
```

```python
selecao = (df["Quartos"] >= 2) & (df["Valor"] < 3000) & (df["Area"] > 70)
df[selecao]
```

```python
df_2 = df[selecao]
```

### **Acrescimo de demanda**

*  Quais apartamentos possuem área maior que 80 m² ou aluguel menor que R$ 4.000,00?

```python
selecao = (df['Area'] > 80) | (df['Valor'] < 4000)
df[selecao]
```
